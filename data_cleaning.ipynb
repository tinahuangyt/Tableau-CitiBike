```python
import pandas as pd
import glob
```

```python
clean_data= pd.concat([pd.read_csv(f) for f in glob.glob('raw_csv/*.csv')], ignore_index = True)
```

```python
clean_data.head()
```

```python
#Replace 0 and 1 with male and female for gender column
clean_data["gender"]=clean_data["gender"].replace(1,"male")
clean_data["gender"]=clean_data["gender"].replace(2,"female")
clean_data["gender"]=clean_data["gender"].replace(0,"na")

#Calculate age from birth year
clean_data["age"]=2019-clean_data["birth year"]

#Convert trip duration to minutes
clean_data['trip duration min']=clean_data["tripduration"]/60

clean_data.head()
```
